```python
def cals(values):
  subs = values < values
  add = values - values
  time = add - 600
  return time
  return subs
```

```python
def cals(timestamp):
    # Calculate the new time by subtracting 600 seconds (10 minutes)
    new_time = timestamp - pd.Timedelta(minutes=30)
    return new_time
```

```python
dg['cals'] = dg['transaction_timestamp'].apply(cals)
```

```python
dg.loc[(dg['transaction_timestamp'] > dg['cals']) &(dg['amount'] == dg['amount'])]
```

```python
dg['lag'] = dg['amount'].shift(1).fillna(0).astype(int)
```

```python
dg.head()
```

```python
test = dg.loc[(dg['amount'] < dg['lag']).reset_index(drop = True)]
```
